1. What is the distribution of sepal lengths and widths among the Iris species?

```python
# KDE plot for Sepal Length
plt.subplot(1, 2, 1)
sns.kdeplot(data=datairis, x="sepal_length", hue="species", fill=True, common_norm=False)
plt.title("Distribution of Sepal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Density")

# KDE plot for Sepal Width
plt.subplot(1, 2, 2)
sns.kdeplot(data=datairis, x="sepal_width", hue="species", fill=True, common_norm=False)
plt.title("Distribution of Sepal Width by Species")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Density")
```

```python
f, ax = plt.subplots(figsize=(11,5))
sns.boxplot(x='sepal_length', y="species",  data=datairis);
```

```python
sns.barplot(x="sepal_width", hue="species", data=datairis);
```
